#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y_x2 = x**2
y_ex = np.exp(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_x2, label='y = x^2', color='red')
plt.plot(x, y_ex, label='y = e^x', color='k')

plt.xscale('linear')
plt.yscale('linear')

plt.figure(figsize=(10, 6))
plt.plot(x, y_x2, label='y = x^2', color='blue')
plt.plot(x, y_ex, label='y = e^x', color='green')

plt.xscale('log')
plt.yscale('log')

plt.show()
```

```python
y_x2 = x**2
y_ex = np.exp(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_x2, label='y = x^2', color='red')
plt.plot(x, y_ex, label='y = e^x', color='k')

plt.xscale('linear')
plt.yscale('linear')

plt.title('Plot of y = x^2 and y = e^x')
plt.xlabel('x')
plt.ylabel('y')

plt.figure(figsize=(10, 6))
plt.plot(x, y_x2, label='y = x^2', color='blue')
plt.plot(x, y_ex, label='y = e^x', color='green')

plt.xscale('log')
plt.yscale('log')

plt.legend()

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED

average_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].mean()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

plt.sca(axes[0])  
average_steps_by_month.plot(kind='bar', ax=axes[0])
plt.title('Average Steps by Month')
plt.xlabel('Month')
plt.ylabel('Average Steps')
plt.xticks(rotation=0)

plt.show()


fitbit['Day_Type'] = fitbit['Work_or_Weekend'].map({0: 'Workday', 1: 'Weekend'})

average_steps_day_type = fitbit.groupby('Day_Type')['Steps'].mean()

plt.sca(axes[1])  
average_steps_day_type.plot(kind='bar', ax=axes[1])
plt.title('Average Steps by Day Type')
plt.xlabel('Day Type')
plt.ylabel('Average Steps')
plt.xticks(rotation=0)

plt.tight_layout()
plt.savefig('figures/average_steps_by_month_and_day_type.png')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

plt.figure(figsize=(10, 2))

for i, (x_feature, y_feature) in enumerate(features, 1):
    plt.subplot(1, 3, i)
    plt.scatter(fitbit[x_feature], fitbit[y_feature], alpha=0.5)
    plt.title(f'{x_feature} vs {y_feature}')
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)

plt.tight_layout()

plt.savefig('figures/scatter_plots.png')

plt.show()
```
