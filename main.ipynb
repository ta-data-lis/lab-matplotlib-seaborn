#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

fig, charts = plt.subplots(nrows=2, ncols=1, figsize=(20,10))

charts[0].set_title('Linear Scale',fontweight='bold')
charts[0].plot(x, y1, 'y--')
charts[0].plot(x, y2, 'g-')

charts[1].set_title('Log Scale')
charts[1].plot(x, y1, 'p--')
charts[1].plot(x, y2, 'r-')

charts[1].set_xscale('log')
charts[1].set_yscale('log')

charts[1].grid(True)
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, charts = plt.subplots(nrows=2, ncols=1, figsize=(20,10))

charts[0].set_title('Linear scale')
charts[0].plot(x, y1, 'y--', label='x^2')
charts[0].plot(x, y2, 'g-', label='exp(x)')

charts[1].set_title('Log scale')
charts[1].plot(x, y1, 'g--', label='x^2')
charts[1].plot(x, y2, 'y-', label='exp(x)')

charts[1].set_xscale('log')
charts[1].set_yscale('log')

charts[0].grid(True)
charts[1].grid(True)
charts[0].legend()
charts[1].legend()

fig.suptitle('Plotting in linear and log scales')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
avg_steps_month = fitbit.groupby('Months_encoded')['Steps'].median()

# MEDIAN STEPS BY WORK_OR_WEEKEND
avg_steps__work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median()


plt.figure(figsize=(12, 4))

# Plot 1: Average number of steps by month
plt.subplot(1, 2, 1)
plt.bar(avg_steps_month.index, avg_steps_month.values)
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.title('Avg Number of Steps by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Plot 2: Average number of steps on weekdays and weekends
plt.subplot(1, 2, 2)
labels = ['Weekend', 'Workday']
plt.pie(avg_steps__work_or_weekend, labels=labels, autopct='%1.1f%%')
plt.title('Median Number of Steps: Workdays vs. Weekends')
plt.savefig('figures/steps_by_month_and_activity.png')

# Show the plots
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
feature_pairs = [('Minutes Lightly Active', 'Steps'),
                 ('Minutes Very Active', 'Steps'),
                 ('Minutes Sedentary', 'Steps')]


plt.figure(figsize=(20, 10))
plt.subplots_adjust(wspace=0.3)  #space between plots
colors = ['red', 'green', 'blue'] 

for i, (x_feature, y_feature) in enumerate(feature_pairs, 1):
    plt.subplot(1, 3, i)
    plt.scatter(fitbit[x_feature], fitbit[y_feature], color=colors[i-1], marker='o', alpha=0.5)
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.title(f'Scatter Plot: {x_feature} vs. {y_feature}')
    plt.grid(True)  # Grid 

plt.savefig('figures/scatter_plots.png')


plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
correlation_matrix = titanic.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

#### What are the most correlated features?
