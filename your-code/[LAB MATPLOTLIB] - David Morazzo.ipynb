#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

montly_steps = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

fig, ax = plt.subplots(1,2, figsize = (12,4))

plt.sca(ax[0])

monthly_steps.plot(kind = 'line', linewidth = 2, color= 'green')

plt.xlabel('Months')
plt.ylabel('Median number of steps')
plt.xticks(range(12), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.title('Median Steps walked by month')

plt.savefig('/Users/R04425/Desktop/my_new_fig.png')


# MEDIAN STEPS BY WORK_OR_WEEKEND

weekend_steps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()

plt.sca(ax[1])

weekend_steps.plot(kind = 'bar', color = 'green', alpha = 0.5)

plt.xlabel('Work_or_Weekend')
plt.ylabel('Median number of steps')
plt.title('Median Steps walked by Workdays and Weekend')
plt.xticks(range(2), ['Weekend','Workdays'], rotation = 0)


plt.savefig('/Users/R04425/Desktop/my_fig.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

```python
# your code here

cols = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for col in cols:
    plt.figure()
    plt.title('Steps vs ' + col)
    plt.xlabel('Steps')
    plt.ylabel(col)
    plt.scatter(fitbit['Steps'], fitbit[col], color= "orange")
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

bins = [10, 20, 50]

f, ax = plt.subplots(1,3)
f.set_figwidth(15)

for i in range(3):
    plt.sca(ax[i])
    titanic['Age'].hist(bins=bins[i])
    plt.title('n =' + str(bins[i]))
    plt.xlabel('Age')
    plt.ylabel('Count')
```

#### How does the bin size affect your plot?

```python
"""
By using a low number of bins we reduce the noise but also the resolution since it becomes less detailed. As we increase the number of bins the resolution increases but becomes more noisy.
"""
```
