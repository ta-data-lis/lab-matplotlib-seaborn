#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
fitbit_monthly_steps = fitbit.groupby('Months_encoded')['Steps'].median()

fitbit_monthly_steps.plot(kind='bar', color='blue', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.title('Average Number of Steps by Month')
plt.xticks(range(0, 12, 1), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# MEDIAN STEPS BY WORK_OR_WEEKEND
fitbit_weekday_steps = fitbit.groupby('Work_or_Weekend')['Steps'].median()

fitbit_weekday_steps.plot(kind='bar', color='green', alpha=0.7)
plt.xlabel('Day Category')
plt.ylabel('Median Steps')
plt.title('Average Number of Steps by Workday vs. Weekend')
plt.xticks([0, 1], ['Workday', 'Weekend'])

# Create the figure and axes
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

plt.tight_layout()
plt.savefig('avg_stps_month_workday_wknd.png')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = [('Minutes Lightly Active', 'Steps'),
            ('Minutes Very Active', 'Steps'),
            ('Minutes Sedentary', 'Steps')]

fig, axes = plt.subplots(1, len(features), figsize=(15, 5))

for i, (x_feature, y_feature) in enumerate(features):
    plt.sca(axes[i])
    plt.scatter(fitbit[x_feature], fitbit[y_feature], c='blue', alpha=0.5)
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.title(f'{x_feature} vs {y_feature}')

plt.tight_layout()
plt.savefig('scatter_plots.png')
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
plt.style.use('classic')
fig, axes = plt.subplots(1, 3, figsize=(12, 6))
bin_values = [10, 20, 50]

for i, bins in enumerate(bin_values):
    ax = axes[i]
    ax.hist(titanic['Age'], bins=bins, color='blue', alpha=0.7)
    ax.set_xlabel('Age')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Age (Bins = {bins})')

plt.tight_layout()
plt.show()
```

#### How does the bin size affect your plot?

```python
"""
The bin size in a histogram determines the width of each bin or interval in which data values are grouped together. 
The choice of bin size can have an impact on the visual representation of the distribution in the histogram.
In the previous example code, three different bin sizes (10, 20, and 50) were used to demonstrate the effect on the 
histogram plots. You can observe that as the bin size increases, the plot becomes smoother but may lose some finer 
details. Conversely, smaller bin sizes provide more detailed information but can introduce more noise.
"""
```
