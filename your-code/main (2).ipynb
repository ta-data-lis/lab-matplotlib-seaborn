#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig, ax = plt.subplots(1, 2, figsize=(10,3))
      
ax[0].plot(x, x**2, 'r', label = 'x**2')
ax[0].plot(x, np.exp(x), 'b', label = 'exp(x)')
ax[0].set_title("Normal scale")


ax[1].plot(x, x**2, 'r', label = 'x**2')
ax[1].plot(x, np.exp(x), 'b', label = 'exp(x)')
ax[1].set_yscale("log")
ax[1].set_xscale("log")
ax[1].set_title("Logarithmic scale (y)")


plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
monthly_steps = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

fig = plt.figure(figsize= (12, 4))
plt.bar(monthly_steps.index, monthly_steps.values, color='green')
plt.xticks(monthly_steps.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.savefig('figures/monthly_steps.png', dpi=1000)

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

weekly_steps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()

fig = plt.figure(figsize= (12, 4))
plt.barh(weekly_steps.index, weekly_steps.values, height= 0.9, color= 'blue', align='center', )
plt.yticks([0, 1], ['Weekend', 'Work'])
plt.savefig('figures/work_or_weekend.png', dpi=1000)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

a='Minutes Lightly Active'
b='Minutes Very Active'
c= 'Minutes Sedentary'

elements = [a, b, c]

for element in elements:
    plt.figure()
    plt.title('Steps vs ' + element)
    plt.xlabel('Steps')
    plt.ylabel(element)
    plt.scatter(fitbit['Steps'], fitbit[element], color="blue")
    
    plt.show()
  
```
